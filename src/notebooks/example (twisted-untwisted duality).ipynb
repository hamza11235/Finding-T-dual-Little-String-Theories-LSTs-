```python
"""
Example showing the duality between twisted and untiwsted LSTs, by realizing them as inequivalent genus-one fibrations within the same CY. 

The duality is checked further by computing matching relevant data in the geometry.
"""


from sage.all import *
load("Quiver.sage")
```

```python
"""
Generating 4D reflexive polytope
"""
in_p=verts([ [-1, -2, 0, 0],   [-1, 2, 0, 0], [1,0,0,0]  , [-1, -2, 2, 0],
          [-1, -1, 2, 0],[-1, 0, 2, 0] ,[-1, 1, 2, 0] ,[-1, 2, 2, 0],
           [-1, -2, -2, 0],
          [-1, -1, -2, 0],[-1, 0, -2, 0] ,[-1, 1, -2, 0] ,[-1, 2, -2, 0],
           [-1, -2,  0, -1],  [-1,-2,0,4],[-1,-2,4,4]
          ])
```

Generating a Reflexive Polytope with at least one genus-one fibration (Note: H:10,92 refers to the Hodge numbers of the CY)

```python
"""
Pick a fibration
"""
p=in_p
num,torus,xproj,yproj,plot=pick_fib(p, 0)
print(num)
```

We find all the inequivalent genus-one fibrations (9 in this case), and pick one of them. We also determine the respective projection vectors for the base of the fibration

```python
"""
Plot 2D subpolytope inside the 3D subpolytope
"""
plot
```

```python
"""
Get CY and base data

Also prints out the non-flat flavor divisors that have been removed
"""
p,letvar, coords, baseindex1, baseindex2, weirs, base=CY_data(p, torus,xproj,yproj)
```

```python
"""
Get base divisors and plot
"""
basedivs=base_dat(base, baseindex1, baseindex2)
```

Note, this is the compact base. To get the LST, we will decompactify the base (remove the ray with both x,y<0)

```python
"""
The quiver for the LST: Field theory data including gauge and flavor algebras, their ranks and Kacs labels.

Coulomb branch dimension also shown. 

This is not the final quiver, as there may be non-flat compact divisors that need to be removed.
"""
basedata(p,base,[],baseindex1,baseindex2)
```

Coulomb branch dimension: 65, Flavor rank: 16 \
The LST quiver for this particular fibration. This is a _twisted_ LST of rank 4. In particular, it is the theory resulting from a $\mathbb{Z}_{2}$ twist of a 6D LST of type $\mathfrak{e}_{6}$, with flavor symmetry $\mathfrak{e}_{8} \times \mathfrak{e}_{8}$

```python
"""
2D base vertices, with their angles from the x-axis
"""
slcomp(undivs(basedivs))
```

```python
"""
From 6D chiral anomalies, one identifies the non-flat compact divisors, and removes them.

checkdivs are base divisors over which the fiber is non-flat, and this is cured by doing a base blow-up, which can be inferred in the final quiver.
"""
checkdivs=[]
p,base,nftotdats,coords,baseindex1,baseindex2,basedivs=rem_nonflat_comp(p, letvar,  base, basedivs,coords, baseindex1,baseindex2,weirs,checkdivs)
```

No non-flat divisors in this case

```python
"""
Computing 2-group structure constants for the LST above (remember to use the intersection data for base after the non-flat resolution).
"""
Two_Group(p,base,basedivs,[],baseindex1,baseindex2)
```

The 2-group data, $(\kappa_{P}, \kappa_{R})$. This, along with the Coulomb branch dimension (given at the end of the quiver) and the flavor rank, is the data we will match across different fibrations, i.e., _T-dual_ LSTs

```python
"""
Pick another fibration
"""
p=in_p
num,torus,xproj,yproj,plot=pick_fib(p, 1)
```

```python
"""
Plot 2D subpolytope inside the 3D subpolytope
"""
plot
```

```python
"""
Get CY and base data
"""
p,letvar, coords, baseindex1, baseindex2, weirs, base=CY_data(p, torus,xproj,yproj)
```

```python
"""
Get base divisors and plot
"""
basedivs=base_dat(base, baseindex1, baseindex2)
```

```python
"""
Quiver for the corresponding LST
"""
basedata(p,base,[],baseindex1,baseindex2)
```

Coulomb branch dimension: 65, Flavor rank: 16 \
The LST quiver for this particular fibration. This is an _untwisted_ LST of rank 2, and type $\mathfrak{e}_{7}$, with flavor symmetry $\mathfrak{so}_{16}\times \mathfrak{so}_{16}$

```python
"""
2D base vertices, with their angles from the x-axis
"""
slcomp(undivs(basedivs))
```

```python
"""
Computing 2-group structure constants for the LST above
"""
Two_Group(p,base,basedivs,[],baseindex1,baseindex2)
```

As we can see, the 2-Group data along with dim(CB) and flavor rank matches that of the first LST!
